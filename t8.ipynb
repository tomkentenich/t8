##TASK-1 translation

```python
df=pd.read_csv('English.csv')
```

```python
df=df.sample(10)
```

```python
def trans(a,b):
  translator=Translator(to_lang=b,from_lang='en')
  return translator.translate(a)
```

```python
df['spanish']=df['English words/sentences'].apply(lambda x: trans(x,'es'))
```

```python
df['french']=df['English words/sentences'].apply(lambda x: trans(x,'fr'))
```

```python
df['german']=df['English words/sentences'].apply(lambda x: trans(x,'de'))
```

```python
import PyPDF2
```

```python
pdffile=open("Resume.pdf",'rb')
```

```python
pdfreader=PyPDF2.PdfFileReader(pdffile)
```

```python
x=pdfreader.getNumPages()
x
```

```python
li=[]
for i in range(x):
  pageobj=pdfreader.getPage(i)
  li.append(pageobj.extractText())
pdffile.close()
```

```python
print(li)
```

```python
li=' '.join(li)
print(li)
li=str(li).lower()
li
```

```python
data_analyst=['data Visualization','data Cleaning','matlab','python','sql','nosql',
'machine learning','linear algebra' , 'Calculus','Microsoft Excel','Critical Thinking','Communication']
```

```python
data_engineer=['Data pipelines', "SQL", 'Python', 'Java',"Hive",'Pig'," Spark", "Hadoop", "MapReduce",
 'Amazon Web Services','AWS', 'MongoDB','Machine learning']
```

```python
datotal=0
for i in data_analyst:
  if len(re.findall(i.lower(),li))!=0:
    for i in re.findall(i.lower(),li):
      datotal=datotal+1
```

```python
datotal
```

```python
detotal=0
for i in data_engineer:
  if len(re.findall(i.lower(),li))!=0:
    for i in re.findall(i.lower(),li):
      detotal=detotal+1
```

```python
detotal
```
